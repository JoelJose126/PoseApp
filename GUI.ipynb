```python
import tkinter as tk
import cv2

class WebcamApp:
    def __init__(self, window, video_source=0):
        self.window = window
        self.window.title("Webcam App")

        self.video_source = video_source

        self.capture = cv2.VideoCapture(self.video_source)

        self.canvas = tk.Canvas(window)
        self.canvas.pack()

        self.delay = 15  # milliseconds
        self.update()

        self.window.mainloop()

    def update(self):
        _, frame = self.capture.read()
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image = tk.PhotoImage(master=self.canvas, width=frame.shape[1], height=frame.shape[0])
        self.canvas.create_image(0, 0, image=image, anchor=tk.NW)
        self.canvas.image = image  # Store reference to prevent garbage collection
        self.canvas.after(self.delay, self.update)

WebcamApp(tk.Tk())

```
